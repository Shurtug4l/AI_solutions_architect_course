# Esempio di gestione dei codici di risposta di un **Endpoint**

```python
%pip install requests

import requests
import json
import random

import requests

# URL base dell'API (esempio)
BASE_URL = "https://cat-fact.herokuapp.com"
# Id da usare nella chiamata API
             #Corretto 200              #Errato 404              #Errore 400
fact_id = ['591f98803b90f7150a19c229', '591f98803b90f7150a19c22A', '0']
#Selezioniamo casualmente uno degli ID
choosen_fact = random.choice(fact_id)

response = requests.get(f"{BASE_URL}/facts/{choosen_fact}")

if response.status_code == 200:
  print("fatto trovato:")
  print(response.json())
elif response.status_code == 404:
  print(f"fatto {choosen_fact} non trovato.")
else:
  print(f"Errore nella richiesta: {response.status_code}")
  print(response.text)
```
