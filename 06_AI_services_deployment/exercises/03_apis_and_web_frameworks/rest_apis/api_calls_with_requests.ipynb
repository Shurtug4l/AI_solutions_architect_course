## POST

```python
url_for_post = 'https://jsonplaceholder.typicode.com/posts'
```

```python
payload = json.dumps(
    {"nome": "Andrea",
     "cognome": "Rossi"}
    )

headers = {
  'Content-Type': 'application/json'
}
```

```python
res = requests.post(url_for_post, data=payload, headers=headers)
```

```python
res.status_code
```

```python
res.json()
```
